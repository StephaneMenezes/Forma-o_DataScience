# Operador de seleção "loc"

```python
#Conseguimos selecionar por dados que queremos, com várias condições: 

#Com uma condição: 

df[['Idade','Nome']].loc[df['Sexo']=='male'].head()
```

```python
#Com varias condições: 

#Com várias condições:

df[['Idade','Nome','Sobreviveu']].loc[(df['Idade']>3) & (df['Idade']<10) & (df['Sexo'] == 'female') \
& (df['Sobreviveu'] == 1)].head()

#Observação importante: Como o código estava ficando muito grande, colcamos uma barra para quebrar o codigo: \ 
```

# Map e replace:

```python
#Substituindo valores nas linhas: com map
df['Sexo'].map({'male':'Homem', 'female':'Mulher'})
```

```python
#Substituindo valores nas linhas: Com replace
df['Sexo'].replace(['female','male'],['Feminino','Masculino'], inplace = True)
```

# Utilizando o comando GroupBy: 

O camando GroupBy assim como no sql serve para agrupar os dados, geralmente por categorias.

```python
df.groupby('Sexo').count()
```

'''Se nós olharmos o resultado acima, nós podemos perceber que a contagem do ID do Passageiro e o Sobreviveu são iguais, isto
acontece, porque ele também está considerado os valores 0 na contagem, para teste vou criar uma várivel utilizando o Loc apenas 
com os valores dos sobreviventes'''

```python
#Armazenando apenas os sobreviventes na variavel: 

sobreviventes = df.loc[df['Sobreviveu'] == 1]
```

```python
#Agora com os dados apenas dos sobreviventes, conseguimos observar que apenas do número de mulheres ser menor que os homens
#O número de mulheres que sobreviveram é bem maior. 
sobreviventes.groupby('Sexo').count()
```

```python
#Agrupando o df pela média de Sexo:

df.groupby('Sexo').mean()
```
